```python
import sentence_transformers
from sentence_transformers import SentenceTransformer
import statistics
```

```python
import nltk
nltk.download('vader_lexicon')
from sklearn.metrics import mean_absolute_error, mean_squared_error
from nltk.sentiment.vader import SentimentIntensityAnalyzer

vader = SentimentIntensityAnalyzer()
```

```python
def cal_average(num):
    sum_num = 0
    for t in num:
        sum_num = sum_num + t           

    avg = sum_num / len(num)
    return avg
```

```python
 def properties(theString,lst):
    print(theString)
    print('min is : ')
    print(min(lst))
    print('max is : ')
    print(max(lst))
    print('median is: ')
    print(statistics.median(map(float, lst)))
    print('average is: ')
    print(cal_average(lst))
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
preds = pd.read_csv('/content/preds.69',  delimiter="\t",  header=None)
test0 = pd.read_csv('/content/reference.test.0',  delimiter="\t",error_bad_lines=False,  header=None)
```

```python
pre= []
for i in range(len(preds)):
    pre.append(preds[0][i])
    
    
```

```python
test0['predictions']= pre
```

```python
test0co = []
for i in range(len(test0)):
    sentences = [test0[0][i], test0['predictions'][i]]
    sentence_embeddings = model.encode(sentences)
    test0co.append(cosine_similarity(
    [sentence_embeddings[0]],
    sentence_embeddings[1:])[0][0])
    
```

```python
test0['BERT_Cosine_Similarity'] = test0co
```

```python
inputFin= test0[0].apply(vader.polarity_scores)
outputFin  = test0['predictions'].apply(vader.polarity_scores)


comp_score_in_Fin= []
for i in range(len(test0)):
    comp_score_in_Fin.append(inputFin[i]['compound'])

test0['Inputs Vader Score'] = comp_score_in_Fin

comp_score_out_Fin= []
for i in range(len(test0)):
    comp_score_out_Fin.append(outputFin[i]['compound'])

test0['Outputs Vader Score'] = comp_score_out_Fin

Fin_diff = []
for i in range(len(test0)):
    Fin_diff.append(test0['Outputs Vader Score'][i] - test0['Inputs Vader Score'][i])

test0['Vader Score Diff'] = Fin_diff
```

**FIND FLIP CLASSES**

```python
flip = []
fromnegativetopositive = 0
fromnegativetoneutral = 0

frompositivetonegative = 0
frompositivetoneutral = 0

fromneutraltonegative = 0
fromneutraltopositive = 0

nochange = 0
for i in range(len(test0)):
    if test0['Inputs Vader Score'][i]<=-0.05:
        
        if test0['Outputs Vader Score'][i]>=0.05:
            flip.append('from negative to positive')
            fromnegativetopositive = fromnegativetopositive +1
        
        elif -0.05<test0['Outputs Vader Score'][i]<0.05:
            flip.append('from negative to neutral')
            fromnegativetoneutral= fromnegativetoneutral +1
        else:
            flip.append('no change')
            nochange= nochange+1

    elif test0['Inputs Vader Score'][i]>=0.05:
        
        if test0['Outputs Vader Score'][i]<=-0.05:
            flip.append('from positive to negative')
            frompositivetonegative= frompositivetonegative+1
        
        elif -0.05<test0['Outputs Vader Score'][i]<0.05:
            flip.append('from positive to neutral')
            frompositivetoneutral = frompositivetoneutral +1
        else:
            flip.append('no change')
            nochange= nochange+1

    else:
        if test0['Outputs Vader Score'][i]<=-0.05:
            flip.append('from neutral to negative')
            fromneutraltonegative = fromneutraltonegative+1
        
        elif test0['Outputs Vader Score'][i]>=0.05:
            flip.append('from neutral to positive')
            fromneutraltopositive = fromneutraltopositive+1
        else:
            flip.append('no change')
            nochange= nochange+1

```

```python
test0['Flip Class'] = flip
```

```python
word_num_list= []
for i in range(len(test0)):
  a_string= test0['predictions'][i]
  word_list = a_string.split()
  number_of_words = len(word_list)
  word_num_list.append(number_of_words)

test0['NumberofWords']=word_num_list
```

```python
frequency = {}

# iterating over the list
for item in word_num_list:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
print(frequency)
```
